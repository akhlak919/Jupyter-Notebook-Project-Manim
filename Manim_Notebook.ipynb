```python
%%manim Fourier
from manim import *

class Fourier(Scene):
    def construct(self):
        math = MathTex(
            "f(x) = \\frac{a_{0}}{2} + \sum_{n = 1}^{\infty} a_{n}\cos\left ( \\frac{2n\pi x}{b-a} \\right )+ \sum_{n = 1}^{\infty} b_{n}\sin\left ( \\frac{2n\pi x}{b-a} \\right )"
        )
        math.set_color_by_gradient(BLUE,RED,GREEN_B,ORANGE, LIGHT_PINK,PURE_GREEN)
        self.add(math)
```

```python
%%manim Sydeny
from manim import *

class Sydeny(Scene):
    def construct(self):
        circle = Circle().set_color(RED_D)
        self.play(Create(circle), run_time=5)
```
